```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/States_and_union_territories_of_India'

response = requests.get(url)

soup = BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
table_part  = soup.find("table",{"class":"wikitable sortable plainrowheaders"})
print(table_part)
```

```python
#extracting only the data in <th> with scope attribute as row
rows = table_part.find_all("th",{"scope":"row"})

#printing the names of states available in the title attribute of <a>
states = []
for state in rows:
    name = state.find('a').get('title')
    states.append(name)
print(states)
```
